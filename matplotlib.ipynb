```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

<h5>
    Além do matplotlib ser uma das primeiras bibliotecas de visualização de dados no python ela é uma base para diversas outras que como o seaborn
</h5>

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
```

<h5>A forma mais simples de se fazer um plot usando o matplotlib é usando o plt.plot(x, y)</h5>

<h5>
    Nesse gráfico temos dois elementos existentes, um deles que é como se fosse uma caixa "invisível" que está ao redor do nosso plot e que vai conter todos os gráficos que colocarmos alí dentro. No matplotlib os gráficos levam o nome de <span style="color:#00ff80">eixo</span> e a caixa imaginária leva o nome de <span style="color:#00ff80">figure</span>.
</h5>

```python
plt.plot(x, y, color="red")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.title("Título")
```

<h5>
    A função subplot permite que criemos dois eixos em uma única figure.
</h5>

```python
plt.subplot(1, 2, 1) # Criando uma figura com 1 linha, 2 colunas e selecionando a primeira
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2) # Criando uma figura com 1 linha, 2 colunas e selecionando a segunda
plt.plot(y, x, 'g*--')
```

<h5>
    Utilizando axes
</h5>

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
ax.plot(x, y, 'b')
ax.set_xlabel("Eixo x")
ax.set_ylabel("Eixo y")
ax.set_title("Título")
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
ax1.plot(x, y, 'b')
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # left, bottom, width, height (range 0 to 1)
ax2.plot(x, y, 'b')
```

<h5>
    A função subplots é muito semelhante a subplot, mas a subplots possui mais customizações.
</h5>

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
ax[0].plot(x, y, 'r--', label="x**2")
ax[0].set_title("Título")
ax[0].legend()
ax[1].plot(y, x, 'r--')
ax[1].set_title("Título")
# plt.tight_layout() Melhora o tamanho da imagem
```
